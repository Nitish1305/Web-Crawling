```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
df = pd.read_excel(r"C:\Users\Nitish.Ranjan\Desktop\Germany\Steffen\Web Crawling\Nitish\MSD\Scrape_MSD_HEPA.xlsx",sheet_name='Sheet2')
```

```python
#df
```

```python
linklist = df["Link"].tolist()
```

```python
#linklist
```

```python
finaltds = []
for link in linklist:
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    mydivs = soup.findAll("div",attrs = {"class": "table-cell search-results-data"})
    tds = []
    for row in mydivs:
        tds.append(row.text)
        finaltds.append(tds)
        tds = []
```

```python
finaltds2 = []
for link in linklist:
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    mydivs = soup.findAll("div",attrs = {"class": "table-cell search-results-specialty"})
    tds = []
    for row in mydivs:
        tds.append(row.text)
        finaltds2.append(tds)
        tds = []
```
